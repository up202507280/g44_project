```python
import sqlite3
import pandas as pd

con = sqlite3.connect('MineDB.db')
cur = con.cursor()

cur.execute('DROP TABLE IF EXISTS "Transaction"')
cur.execute('DROP TABLE IF EXISTS "Inspection"')
cur.execute('DROP TABLE IF EXISTS "Site"')
cur.execute('DROP TABLE IF EXISTS "Company"')
cur.execute('DROP TABLE IF EXISTS "Category"')

command = 'CREATE TABLE "Category" ("id" INTEGER,'
command += '"name" TEXT UNIQUE, '
command += 'PRIMARY KEY("id" AUTOINCREMENT))'
cur.execute(command)

command = 'CREATE TABLE "Company" ("id" INTEGER,'
command += '"name" TEXT NOT NULL, "created_date" TEXT, '
command += 'PRIMARY KEY("id"))'
cur.execute(command)

command = 'CREATE TABLE "Site" ("id" INTEGER,'
command += '"tittle" TEXT, "category_id" INTEGER, '
command += 'PRIMARY KEY("id"), '
command += 'FOREIGN KEY("category_id") REFERENCES "Category"("id"))'
cur.execute(command)

command = 'CREATE TABLE "Inspection" ("id" INTEGER,'
command += '"info" TEXT, "company_id" INTEGER, '
command += 'PRIMARY KEY("id"))'
cur.execute(command)

command = 'CREATE TABLE "Transaction" ("id" INTEGER,'
command += '"date" TEXT, "amount" REAL, "company_id" INTEGER, "site_id" INTEGER, '
command += 'PRIMARY KEY("id" AUTOINCREMENT))'
cur.execute(command)

df = pd.read_csv('g44_Mining_Sites_v2.csv')

categories = df[['site_category']].dropna().drop_duplicates()
for val in categories['site_category']:
    cur.execute('INSERT INTO Category (name) VALUES (?)', (val,))

companies = df[['company_id', 'company_name']].drop_duplicates()
for _, row in companies.iterrows():
    cur.execute('INSERT OR IGNORE INTO Company (id, name) VALUES (?, ?)', (row['company_id'], row['company_name']))

cat_df = pd.read_sql('SELECT * FROM Category', con)
sites_df = df[['site_id', 'site_title', 'site_category']].drop_duplicates()
sites_final = sites_df.merge(cat_df, left_on='site_category', right_on='name')

for _, row in sites_final.iterrows():
    cur.execute('INSERT INTO Site (id, tittle, category_id) VALUES (?, ?, ?)', (row['site_id'], row['site_title'], row['id']))

for _, row in df.iterrows():
    cur.execute('INSERT INTO "Transaction" (date, amount, company_id, site_id) VALUES (?, ?, ?, ?)', (row['transaction_date'], row['amount'], row['company_id'], row['site_id']))

con.commit()
con.close()
print("MineDB.db criada.")
```
